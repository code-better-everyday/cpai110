```python
def shopping_cart_total(items, discount_percent=0):
    """Compute the total cost of a shopping cart with an optional discount.

    Args:
        items: list of (name, price, quantity) tuples.
        discount_percent: percentage discount to apply (0-100).

    Returns:
        A dict with the subtotal, discount amount, and final total.
    """
    if not 0 <= discount_percent <= 100:
        raise ValueError("discount_percent must be between 0 and 100")

    subtotal = sum(price * quantity for _, price, quantity in items)
    discount_amount = subtotal * (discount_percent / 100)
    total = subtotal - discount_amount

    return {
        "subtotal": round(subtotal, 2),
        "discount_amount": round(discount_amount, 2),
        "total": round(total, 2),
    }


# Example usage
cart = [
    ("apple", 0.50, 6),
    ("bread", 2.25, 2),
    ("milk", 1.80, 1),
]

result = shopping_cart_total(cart, discount_percent=50)
print(result)
```
